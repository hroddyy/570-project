4. Test on 50 audio samples not used for training retrieved from google drive to attain metrics to determine the success of the model

```python

import os
import dac
from audiotools import AudioSignal
import torch
import numpy as np
from scipy.io import wavfile
import gc
import csv

# Function to compute Mean Squared Error (MSE)
def compute_mse(original, reconstructed):
    # Ensure both signals are of the same length
    min_length = min(len(original), len(reconstructed))
    return np.mean((original[:min_length] - reconstructed[:min_length]) ** 2)

# Function to compute Signal-to-Noise Ratio (SNR)
def compute_snr(original, reconstructed):
    mse = compute_mse(original, reconstructed)
    return 10 * np.log10(np.var(original) / mse)

# Download and load a model
model_path = dac.utils.download(model_type="44khz")
model = dac.DAC.load(model_path, weights_only=True)  # Set weights_only=True

# Move model to GPU if available
model.to('cuda')

def process_and_collect_metrics(start_file_num, num_files, audio_dir, output_file):
   with open(output_file, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["File Number", "MSE", "SNR"])  # Write header row


    for i in range(num_files):
      file_num = start_file_num + i
      file_path = os.path.join(audio_dir, f"{file_num}.wav")

      if os.path.exists(file_path):
        # Load audio signal file on CPU
        signal = AudioSignal(file_path)

        # Preprocess the audio signal on CPU
        with torch.no_grad():
            chunk_size = 16384  # Adjust this value as needed
            num_chunks = int(np.ceil(signal.audio_data.shape[-1] / chunk_size))

            x_chunks = []
            for j in range(num_chunks):
                start = j * chunk_size
                end = min((j + 1) * chunk_size, signal.audio_data.shape[-1])
                chunk = signal.audio_data[:, :, start:end]
                x_chunks.append(model.preprocess(chunk, signal.sample_rate))  # Assuming preprocess can run on CPU

            x = torch.cat(x_chunks, dim=-1)

            # Move data to GPU for encoding and decoding
            #model.to('cuda')
            x = x.to('cuda')

            # Encode and decode
            z, codes, latents, _, _ = model.encode(x)
            y = model.decode(z)

            # Move necessary data back to CPU
            codes = codes.cpu()
            y = y.cpu()

            # Compress and save before loading
            compressed_signal = model.compress(signal)
            compressed_signal.codes = compressed_signal.codes.cpu()
            compressed_signal.input_db = compressed_signal.input_db.cpu()
            compressed_signal.save(f"compressed_{file_num}.dac") # Save the compressed file

            # Decompress
            loaded_signal = dac.DACFile.load(f"compressed_{file_num}.dac")  # Now load the saved file
            reconstructed_signal = model.decompress(loaded_signal)
            reconstructed_signal.cpu()

            # Move model back to CPU after processing each file
            #model.cpu()
            #torch.cuda.empty_cache()

        # ... (metric calculation and saving, ensure data is on CPU) ...
        reconstructed_audio_data = reconstructed_signal.audio_data.cpu().numpy()

        # Load original audio for metric computation
        sample_rate_orig, original_audio_data = wavfile.read(file_path)

        # Compute metrics
        mse_value = compute_mse(original_audio_data, reconstructed_audio_data)
        snr_value = compute_snr(original_audio_data, reconstructed_audio_data)
        writer.writerow([file_num, mse_value, snr_value])


        # Explicitly delete variables and run garbage collection
        del signal, x, z, codes, latents, y, compressed_signal, loaded_signal, reconstructed_signal
        gc.collect()
        torch.cuda.empty_cache()

      else:
        print(f"Skipping file {file_path} (does not exist)")



# Example usage:
start_file_num = 338
num_files = 50
audio_dir = "/content/drive/MyDrive/ECE570/code/ARCA23K/ARCA23K.train"  # Update with your audio directory
output_file = "/content/drive/MyDrive/ECE570/code/ARCA23K/metrics.csv"  # Specify output file path
process_and_collect_metrics(start_file_num, num_files, audio_dir, output_file)

# ... (further processing of the collected metrics) ...
```

5. Attain metrics and associated graphs

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load metrics from CSV file
csv_file_path = '/content/drive/MyDrive/ECE570/code/ARCA23K/metrics.csv'  # Update this path
df = pd.read_csv(csv_file_path)

# Descriptive Statistics
print("Descriptive Statistics:")
print(df[['MSE', 'SNR']].describe())

# Box Plot for MSE
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data=df['MSE'])
plt.title('Box Plot of Mean Squared Error (MSE)')  # Title added here
plt.yscale('log')  # Use log scale if MSE varies widely
plt.ylabel('Mean Squared Error (MSE)')  # Y-axis label

# Box Plot for SNR
plt.subplot(1, 2, 2)
sns.boxplot(data=df['SNR'])
plt.title('Box Plot of Signal-to-Noise Ratio (SNR)')  # Title added here
plt.ylabel('Signal-to-Noise Ratio (SNR)')  # Y-axis label

plt.tight_layout()
plt.show()

# Scatter Plot of MSE vs SNR
plt.figure(figsize=(8, 6))
sns.scatterplot(x='MSE', y='SNR', data=df)
plt.title('Scatter Plot of MSE vs SNR')  # Title added here
plt.xscale('log')  # Use log scale for better visualization if needed
plt.xlabel('Mean Squared Error (MSE)')  # X-axis label
plt.ylabel('Signal-to-Noise Ratio (SNR)')  # Y-axis label
plt.show()

# Identify problematic files with high MSE or low SNR
high_mse_files = df[df['MSE'] > df['MSE'].quantile(0.75)]
low_snr_files = df[df['SNR'] < df['SNR'].quantile(0.25)]

print("Files with High MSE:")
print(high_mse_files)

print("Files with Low SNR:")
print(low_snr_files)
```
